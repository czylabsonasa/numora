```python
import numpy as np
import scipy.linalg as scl
eigen=scl.eigh
import numpy.random as nr
import sympy as sp
import matplotlib.pyplot as plt

import time
```

```python
stime=time.time()
A=np.matrix([[1.0, 2.0], [-1.0, 3.]])
#A/=np.linalg.norm(A,1)
#print(np.linalg.norm(A,1))
step=5
cc=[str(i) for i in np.linspace(0.1,0.9,step)]
x=np.array(np.zeros(2*step),order='F').reshape(2,step)
#print(x)
N=50
plt.figure(figsize=(15, 15))
plt.axis('equal')
for it in np.linspace(0,2*np.pi,N):
    x[:,0]=np.array([np.cos(it),np.sin(it)])
    for i in np.arange(1,step):
        x[:,i]=np.dot(A,x[:,i-1])
    plt.scatter(x[0],x[1],color=cc,s=1)
    for i in np.arange(step-1):
        a=plt.arrow(x[0,i],x[1,i],
                  0.8*(x[0,i+1]-x[0,i]),0.8*(x[1,i+1]-x[1,i]),
                    color=cc[i])
print(time.time()-stime)
print(eigen(A))
plt.show()
```
