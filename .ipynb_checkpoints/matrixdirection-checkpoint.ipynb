```python
import numpy as np
import scipy.linalg as scl
eigen=scl.eigh
import numpy.random as nr
import sympy as sp
import matplotlib.pyplot as plt
```

```python
A=np.matrix([[1.0, 2], [-2.0,1]])
A/=np.linalg.norm(A,1)
#print(np.linalg.norm(A,1))
numpts=10
cc=[str(1.0-i) for i in np.linspace(0.2,0.9,numpts)]
x=np.array(np.zeros(2*numpts)).reshape(2,numpts)
#print(x)
N=36
for it in np.linspace(0,2*np.pi,N):
    x[:,0]=np.array([np.cos(it),np.sin(it)])
    for i in np.arange(1,numpts):
        x[:,i]=np.dot(A,x[:,i-1])
    plt.scatter(x[0],x[1],color=cc,s=1)
    for i in np.arange(numpts-1):
        plt.arrow(x[0,i],x[1,i],
                  x[0,i+1]-x[0,i],x[1,i+1]-x[1,i],
                  shape='left',width=0.0001,animated=True)
plt.show()
```
